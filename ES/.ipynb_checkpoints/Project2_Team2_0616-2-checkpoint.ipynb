```python
import gensim
import pandas as pd
import jieba
import jieba.posseg as pseg
from collections import defaultdict, Counter
```

```python
NC_excel = pd.read_excel("NC_1_All.xlsx")
NC_Content_List = NC_excel.News_Content.astype(str).tolist()
NC_Content_List_Join = "".join(NC_Content_List)
```

```python
documents = []
for i in NC_Content_List:
    documents.append((jieba.lcut(i)))
```

```python
print(documents[0])
```

```python
model = gensim.models.Word2Vec (documents, size=150, window=10, min_count=2, workers=10)
#model.train(documents,total_examples=len(documents),epochs=10)
```

```python
def get_words(txt, most_common_N=5):
    most_common_verb = []
    seg_list = jieba.cut(txt)
    c = Counter()
    for x in seg_list:
        if len(x)>1 and x != '\r\n':
            c[x] += 1
    for (k,v) in c.most_common(most_common_N):
        most_common_verb.append(k)
    return most_common_verb
most_common_words = get_words(NC_Content_List_Join, most_common_N=100)
with open("most_common_words.txt", "wb") as fp:   #Pickling
    pickle.dump(most_common_words, fp)
"""with open("most_common_words.txt", "rb") as fp:   # Unpickling
    most_common_words = pickle.load(fp)"""
```

```python
# most common verblist
import copy
most_common_words_remain=copy.deepcopy(most_common_words)

sent = pseg.cut(NC_Content_List_Join.replace("。", "")) 
verblist = []
for i, (word, flag) in enumerate(sent):
    if i % 100 ==0:
        print(i)
    if flag == 'v':
        if word in most_common_words_remain:
            verblist.append(word)
            most_common_words_remain.remove(word)
    if len(verblist)>=5:
        break
verblist
```

```python
# most common nounlist
import copy
most_common_words_remain=copy.deepcopy(most_common_words)

sent = pseg.cut(NC_Content_List_Join.replace("。", "")) 
nounlist = []
for i, (word, flag) in enumerate(sent):
    if i % 100 ==0:
        print(i)
    if flag == 'n':
        if word in most_common_words_remain:
            nounlist.append(word)
            most_common_words_remain.remove(word)
    if len(nounlist)>=5:
        break
nounlist
```

```python
for i in verblist:
    print('\n與「%s」最相似的20個詞彙:' % i)
    w = model.wv.most_similar (positive=i,topn=20)
    for i in w :
        print('%s:%.3f   ' % (i[0],i[1]),end='')
```

```python
for i in nounlist:
    print('\n與「%s」最相似的20個詞彙:' % i)
    w = model.wv.most_similar (positive=i,topn=20)
    for i in w :
        print('%s:%.3f   ' % (i[0],i[1]),end='')
```
