## 31個城市的座標

```python
data=[(1304,2312),(3639,1315),(4177,2244),(3712,1399),
(3488,1535),(3326,1556),(3238,1229),(4196,1004),
(4312,790),(4386,570),(3007,1970),(2562,1756),
(2788,1491),(2381,1676),(1332,695),
(3715,1678),(3918,2179),(4061,2370),
(3780,2212),(3676,2578),(4029,2838),
(4263,2931),(3429,1908),(3507,2367),
(3394,2643),(3439,3201),(2935,3240),
(3140,3550),(2545,2357),(2778,2826),
(2370,2975)]
```

## 定義目標函數

```python
def cj(ki):
    sum = 0
    for i in range(0,30):
        sum = sum +  math.sqrt(pow((data[ki[i]][0]-data[ki[i+1]][0]),2) + pow((data[ki[i]][1]-data[ki[i+1]][1]),2))
        # 距離 ((x1 - x2)^2 + (y1 - y2)^2) 的平方根 
    return sum
```

## 初始解生成 (城市排的順序)

```python
xold=[]
for i in range(0,31):
    xold.append(i)
xnew=xold[:]
```

## 溫度參數 以及 生成鄰近解的函數設定

```python
#tmax是初始最高温度，tmin是最低温度，cx是溫度係數
tmax=30000
tmin=1e-8
cx=0.98
#count用來計算新解連續"沒有"被採用的次數
count=0
#get_new_jie用來生成新的鄰近解
def get_new_jie(xold):
    a=random.randint(0,30)
    b=random.randint(0,30)

    tempold=xold[:]

    temp=tempold[a]
    tempold[a]=tempold[b]
    tempold[b]=temp
    tempnew=tempold
    return tempnew
```

```python
get_new_jie(xnew)
```

## SA退火過程

```python
while(tmax>tmin):
    #在tmax温度下找到最好的解xnew
    for i in range(1,5000):
        tempnew=get_new_jie(xold) # 產生鄰近解
        de=cj(tempnew)-cj(xold) # 與原始解相減
        if de < 0 : 
            count=1 # 被採用
            xold=xnew 
            xnew=tempnew # 更新最佳順序

        else:
            count=count+1 # 沒有被採用
            d=pow(math.e,-de/tmax) # 溫度函數計算 e ^ (-f(x)/T)
            if random.random()<d : # 判斷是否大於隨機數
                xold = xnew
                xnew = tempnew
            else:
                pass
    print(cj(xnew)) # 當下最佳解的值
    tmax = tmax * cx  #退火
    #如果连续5000次以上新解没有被接受，则说明已有解xold已经足够优，因此可以退出退火
    if count>5000 :
        break
print(count)
print(xnew)
print(cj(xnew))
```
